```python
import pandas as pd

df = pd.read_csv("train.csv")
```

```python
df.columns = df.columns.str.strip()
```

```python
import numpy as np

base_fare = {
    'Eco': 50,
    'Eco Plus': 80,
    'Business': 150
}

travel_coef = {
    'Business travel': 1.2,
    'Personal Travel': 1.0
}

distance_rate = 0.15
```

```python
df['Price'] = (
    df['Class'].map(base_fare) +
    df['Flight Distance'] * distance_rate
) * df['Type of Travel'].map(travel_coef)

# Добавляем шум
df['Price'] += np.random.normal(0, 20, len(df))

# Защита от отрицательных значений
df['Price'] = df['Price'].clip(lower=30)


df['Price'] = df['Price'].round(2)
```

```python
df['Price'].describe()
```

```python
df.groupby('Class')['Price'].mean()
```

```python
df.to_csv("train_with_price.csv", index=False)
```
